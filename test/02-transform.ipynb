```python
import pandas as pd
```

### Transform province and population data

```python
df_prov_pop = pd.read_csv("extracted_data/province_population.csv")
df_prov_pop.head()
```

```python
df_prov_pop.value_counts()
```

```python
df_prov_pop['Province'].unique()
```

```python
df_prov_pop = df_prov_pop[df_prov_pop.Province != "Philippines"]
df_prov_pop['Province'].unique()
```

```python
df_prov_pop.dtypes
```

```python
df_prov_pop['Population'] = df_prov_pop['Population'].str.replace(',','')

df_prov_pop.head()
```

```python
df_prov_pop.dtypes
```

```python
df_prov_pop["Population"] = df_prov_pop["Population"].astype('int')
```

```python
df_prov_pop.dtypes
```

```python
df_prov_pop.columns = [x.lower() for x in df_prov_pop.columns]
df_prov_pop.columns = df_prov_pop.columns.str.replace("[ ]", "_", regex=True)
df_prov_pop.head()
```

```python
df_prov_pop.to_csv("transformed_data/province_population.csv", index = False)
```

### Transform venues data

```python
df_venues = pd.read_csv("extracted_data/province_venues.csv")
df_venues.tail()
```

```python
df_venues.columns = [x.lower() for x in df_venues.columns]
df_venues.columns = df_venues.columns.str.replace("[ ]", "_", regex=True)
df_venues.tail()
```

```python
df_venues = df_venues[df_venues.province != "Philippines"]
df_venues.tail()
```

```python
df_venues.dtypes
```

```python
df_quezon = df_venues[df_venues.province == "Quezon"]
df_quezon.head()
```

```python
df_quezon.province = "Metro Manila"
df_quezon.head()
```

```python
df_venues_qc_filtered = df_venues[df_venues.province != "Quezon"]
```

```python
df_venues_transformed = pd.concat([df_venues_qc_filtered, df_quezon])
df_venues_transformed.tail(100)
```

```python
df_venues_transformed.province.unique()
```

```python
df_venues_transformed.to_csv("transformed_data/province_venues.csv", index = False)
```

### Transform province tourism data

```python
df_tourists = pd.read_csv("extracted_data/province_tourism.csv")
df_tourists.head(50)
```

```python
df_tourists.columns = [x.lower() for x in df_tourists.columns]
df_tourists.columns = df_tourists.columns.str.replace("[ ]", "_", regex=True)
df_tourists.head()
```

```python
df_tourists.isnull().sum()
```

```python
df_tourists["no._of_foreign_tourists"].value_counts()
```

```python
df_tourists = df_tourists[df_tourists["no._of_foreign_tourists"] != "-"]
df_tourists.head(50)
```

```python
df_tourists.dtypes
```

```python
df_tourists['no._of_foreign_tourists'] = df_tourists['no._of_foreign_tourists'].str.replace(',','')
df_tourists["no._of_foreign_tourists"] = df_tourists["no._of_foreign_tourists"].astype('int')
df_tourists.dtypes
```

```python
df_tourists.head(20)
```

```python
df_tourists.nunique()
```

```python
df_tourists.rename(columns = {"no._of_foreign_tourists" : "num_tourists"}, inplace = True)
df_tourists.head()
```

```python
df_tourists.to_csv("transformed_data/province_tourists.csv", index=False)
```
