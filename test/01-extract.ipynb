### Extract province and population data from web

```python
# Import libraries
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# Define data source url
url = "https://en.wikipedia.org/wiki/List_of_Philippine_provinces_by_population"

# Obtain source text
page = requests.get(url).text
print(page)
```

```python
# Parse html page with Beautiful Soup
data = BeautifulSoup(page,'html.parser')
print(data)
```

```python
# Get all table body elements by finding the tbody HTML tag
tables = data.find_all('tbody')
print(tables)
```

```python
# Get all row elements by finding the tr HTML tag from the first table
rows = tables[0].find_all('tr')
print(rows)
```

```python
# Get all data cell elements by finding the td HTML tag from each row
for row in rows:
    col = row.find_all('td')
    print(col)
```

```python
# Exclude empty col
for row in rows:
    col = row.find_all('td')
    if (col != []):
        # Print cell values for first column
        print(col[0]) # Change this to 1 to get second column cells
```

```python
# Check error cause
for row in rows:
    col = row.find_all('td')
    if (col != []):
        # Print the number of columns per row
        print(len(col)) # Last row of the table has only 1 column causing the IndexError: list index out of range 
```

```python
# Ensure to only get columns where there are at least 2 columns
for row in rows:
    col = row.find_all('td')
    if (col != []):
        if len(col) > 1:
            print(col[1]) # No errors
```

```python
# Get province and population using .text
for row in rows:
    col = row.find_all('td')
    if (col != []):
        if len(col) > 1:
            print(col[0].text)
            print(col[1].text)
```

```python
# Use strip and replace methods for province to remove trailing charactes at the end of the string
for row in rows:
    col = row.find_all('td')
    if (col != []):
        if len(col) > 1:
            print(col[0].text.strip().replace('[a]', ''))
            print(col[1].text)
```

```python
# Import pandas library
import pandas as pd
```

```python
# Create the province and population dataframe
df_prov_pop = pd.concat([pd.DataFrame([{"Province": col[0].text.strip().replace('[a]', ''),
                               "Population": col[1].text}])
                for row in rows if (col := row.find_all('td')) and col != [] and len(col) > 1],
               ignore_index=True)

df_prov_pop.head()
```

```python
df_prov_pop.to_csv("extracted_data/province_population.csv", index=False)
```

### Extract coordinates data from Geocoder

```python
# Import geocoder library
import geocoder
```

```python
# Make a list of the provinces
provinces_list = df_prov_pop["Province"].tolist()
print(provinces_list)
```

```python
# Create an empty list
coordinates = []

# Loop into the list of provinces
for province in provinces_list:
    # Get geocode address of each province
    g = geocoder.arcgis('{}, Philippines'.format(province))
    print(g)
```

```python
# Loop into the list of provinces
for province in provinces_list:
    g = geocoder.arcgis('{}, Philippines'.format(province))
    # Get latiture and longitude of each geocode address
    latlng = g.latlng
    print(latlng)
```

```python
for province in provinces_list:
    g = geocoder.arcgis('{}, Philippines'.format(province))
    latlng = g.latlng
    # Append the obtained longitude and latitude to the empty coordinates list
    coordinates.append(latlng)

print(coordinates)
```

```python
# Convert coordinates list to a dataframe
df_coordinates = pd.DataFrame(coordinates, columns=['Latitude', 'Longitude'])
df_coordinates.head()
```

```python
df_coordinates.to_csv("extracted_data/province_coordinates.csv", index=False)
```

### Extract venues data from Foursquare API

```python
# Make a series of province name and their corresponding coordinates
province_names = df_prov_pop["Province"]
coordinates = [(lat, lng) for lat, lng in zip(df_coordinates["Latitude"], df_coordinates['Longitude'])]

print(province_names)
print(coordinates)
```

```python
# Import os library
import os
```

```python
# Loop into the coordinates and provinces
for (lat, lng), name in zip(coordinates, province_names):
    
    # FourSquare url
    url = "https://api.foursquare.com/v3/places/search"

    # Qeury venues data based on latitude and longitude, arranged by distance
    params = {
        "query": "venue",
        "ll": f"{lat},{lng}",
        "sort": "DISTANCE",
    }

    # Authorization
    headers = {
        "Accept": "application/json",
        "Authorization": os.environ.get("API_KEY"),
    }

    # Get venues data for each coordinate
    response = requests.get(url, params=params, headers=headers)

    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()
        print(data) # Copy to a notepad
```

```python
import os
import pandas as pd
import requests

# Create empty lists to store data
province_data = []
venue_data = []

# Loop through the JSON responses
for (lat, lng), province_name in zip(coordinates, province_names):
    url = "https://api.foursquare.com/v3/places/search"
    params = {
        "query": "venue",
        "ll": f"{lat},{lng}",
        "radius": '22000',
        # "sort": "DISTANCE",
    }
    headers = {
        "Accept": "application/json",
        "Authorization": os.environ.get("API_KEY"),
    }
    
    response = requests.get(url, params=params, headers=headers)

    if response.status_code == 200:
        data = response.json()
        
        # Extract venue information
        for venue in data.get('results', []):
            venue_info = {
                'Province': province_name,
                'Venue Name': venue['name'],
                'Venue Category': venue.get('categories', [])[0].get('name'),
                'Venue Lat': venue['geocodes']['main']['latitude'],
                'Venue Long': venue['geocodes']['main']['longitude'],
                'Address': venue['location']['formatted_address']
            }
            venue_data.append(venue_info)

# Create DataFrame
df_venue = pd.DataFrame(venue_data)

# Display DataFrames
print("\nVenue DataFrame:")
df_venue.head(100)
```

### Extract province geometry coordinates from geojson

```python
geo_url = "https://raw.githubusercontent.com/macoymejia/geojsonph/master/Province/Provinces.json"

try:
    response = requests.get(geo_url)
    if response.status_code == 200:
        with open("extracted_data/province_geometry.json", 'wb') as file:
            file.write(response.content)
        print(f'GeoJSON file downloaded!')
    else:
        print(f'Failed to download GeoJSON. Status code: {response.status_code}')
except Exception as e:
    print(f'An error occurred: {e}')
```
