### Extract province and population data

```python
# Import libraries
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# Define data source url
url = "https://en.wikipedia.org/wiki/List_of_Philippine_provinces_by_population"

# Obtain source text
page = requests.get(url).text
print(page)
```

```python
# Parse html page with Beautiful Soup
data = BeautifulSoup(page,'html.parser')
print(data)
```

```python
# Get all table body elements by finding the tbody HTML tag
tables = data.find_all('tbody')
print(tables)
```

```python
# Get all row elements by finding the tr HTML tag from the first table
rows = tables[0].find_all('tr')
print(rows)
```

```python
# Get all data cell elements by finding the td HTML tag from each row
for row in rows:
    col = row.find_all('td')
    print(col)
```

```python
# Exclude empty col
for row in rows:
    col = row.find_all('td')
    if (col != []):
        # Print cell values for first column
        print(col[0]) # Change this to 1 to get second column cells
```

```python
# Check error cause
for row in rows:
    col = row.find_all('td')
    if (col != []):
        # Print the number of columns per row
        print(len(col)) # Last row of the table has only 1 column causing the IndexError: list index out of range 
```

```python
# Ensure to only get columns where there are at least 2 columns
for row in rows:
    col = row.find_all('td')
    if (col != []):
        if len(col) > 1:
            print(col[1]) # No errors
```

```python
# Get province and population using .text
for row in rows:
    col = row.find_all('td')
    if (col != []):
        if len(col) > 1:
            print(col[0].text)
            print(col[1].text)
```

```python
# Use strip and replace methods for province to remove trailing charactes at the end of the string
for row in rows:
    col = row.find_all('td')
    if (col != []):
        if len(col) > 1:
            print(col[0].text.strip().replace('[a]', ''))
            print(col[1].text)
```

```python
import pandas as pd
```

```python
# Create an empty province and population dataframe
df = pd.concat([pd.DataFrame([{"Province": col[0].text.strip().replace('[a]', ''),
                               "Population": col[1].text}])
                for row in rows if (col := row.find_all('td')) and col != [] and len(col) > 1],
               ignore_index=True)

df.head()
```
