```python
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Philippine_provinces_by_population'
```

```python
attribute_list = ["Province", "Population"]  
```

```python
page = requests.get(url).text
data = BeautifulSoup(page,'html.parser')
df = pd.DataFrame(columns=attribute_list)
table = data.find('table',{'class':'wikitable sortable'})

print(table)
```

```python
for row in table.find_all('a'):
    col = row.get('title')
    print(col)
```

```python
page = requests.get(url).text
data = BeautifulSoup(page,'html.parser')
df = pd.DataFrame(columns=attribute_list)
tables = data.find_all('tbody')
rows = tables[0].find_all('tr')
```

```python
for row in rows:
    col = row.find_all('td')
    if len(col) > 1:
        province = col[1].text
        print(province)
```
